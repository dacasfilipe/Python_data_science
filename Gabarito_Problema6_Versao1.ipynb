```python
# Dicionário para armazenar os resultados dos jogos
resultados_jogos = {}

# Variáveis para contagem das vitórias, empates e saldo de gols de cada time
vitorias_avai = 0
vitorias_figueirense = 0
empates = 0
saldo_gols_avai = 0
saldo_gols_figueirense = 0

# Loop para entrada dos resultados dos jogos
for jogo in range(1, 6):
    partida = input(f"Digite o resultado do jogo {jogo} (Avaí x Figueirense): ").upper()
    placar1, placar2 = partida.split("X")
    resultado_avai = int(placar1)
    resultado_figueirense = int(placar2)

    resultados_jogos[jogo] = {
        'Avaí': resultado_avai,
        'Figueirense': resultado_figueirense
    }

    if resultado_avai > resultado_figueirense:
        vitorias_avai += 1
    elif resultado_figueirense > resultado_avai:
        vitorias_figueirense += 1
    else:
        empates += 1

    saldo_gols_avai += resultado_avai - resultado_figueirense
    saldo_gols_figueirense += resultado_figueirense - resultado_avai

# Exibe as informações dos jogos
print("\n=== Resultados dos Jogos ===")
for jogo, resultados in resultados_jogos.items():
    resultado_avai = resultados['Avaí']
    resultado_figueirense = resultados['Figueirense']
    print(f"Jogo {jogo} (Avaí x Figueirense): {resultado_avai} x {resultado_figueirense}")

# Exibe as informações finais
print(resultados_jogos)
print("\n=== Estatísticas ===")
print("Quantidade de vitórias do Avaí:", vitorias_avai)
print("Quantidade de vitórias do Figueirense:", vitorias_figueirense)
print("Quantidade de empates:", empates)
print("Saldo de gols do Avaí:", saldo_gols_avai)
print("Saldo de gols do Figueirense:", saldo_gols_figueirense)

```
