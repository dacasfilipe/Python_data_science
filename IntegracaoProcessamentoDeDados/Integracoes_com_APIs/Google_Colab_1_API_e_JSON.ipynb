## <h1> Dados Estruturados </h1>

Os dados estruturados são organizados em uma estrutura predefinida e bem definida, onde as relações entre os campos são claras e fixas. Geralmente, esses dados são armazenados em tabelas de bancos de dados ou em formatos de arquivo tabulares, como CSV (Comma-Separated Values) e planilhas Excel. Em Python, a biblioteca pandas é amplamente usada para manipular e analisar dados estruturados.



Exemplo de dados estruturados em Python usando a biblioteca pandas:



```python
import pandas as pd

# Criar um DataFrame a partir de um dicionário
data = {
    'Nome': ['Alice', 'Bob', 'Carol'],
    'Idade': [25, 30, 28],
    'Cidade': ['São Paulo', 'Nova York', 'Los Angeles']
}

df = pd.DataFrame(data)
print(df)
```

```python
print(data)
```

```python
import json

# Dados em formato JSON
dados_json = '{"nome": "Carlos", "idade": 30, "cidade": "Rio de Janeiro"}'

# Converter JSON para dicionário Python
dados_dict = json.loads(dados_json)

print(dados_dict['nome'])
print(dados_dict['idade'])
```

```python
print(dados_json)
```

```python
import xml.etree.ElementTree as ET

# Dados em formato XML
dados_xml = '''<pessoa>
    <nome>Julia</nome>
    <idade>25</idade>
    <cidade>São Paulo</cidade>
</pessoa>'''

# Analisar o XML
raiz = ET.fromstring(dados_xml)

nome = raiz.find('nome').text
idade = raiz.find('idade').text

print(f"Nome: {nome}, Idade: {idade}")
```

```python
print(dados_xml)
```

## <h1> Dados não-estruturados </h1>
Os dados não estruturados não têm uma estrutura fixa e organizada. Eles podem ser informações de texto, áudio, vídeo, imagens ou qualquer formato que não se encaixe em uma estrutura tabular tradicional. Em Python, você pode lidar com dados não estruturados usando várias bibliotecas, dependendo do tipo de dados, como NLTK (processamento de linguagem natural), librosa (áudio) ou OpenCV (imagens e vídeos).


Exemplo de dados não estruturados em Python:



```python
# Exemplo de análise de texto usando NLTK
import nltk
from nltk.tokenize import sent_tokenize, word_tokenize

nltk.download('punkt')

texto = "O sol brilha no céu azul. As aves voam livres."

palavras = word_tokenize(texto)
print(palavras)

sentencas = sent_tokenize(texto)
print(sentencas)
```

Em resumo, a principal diferença entre dados estruturados e não estruturados em Python está na organização e formato dos dados. Dados estruturados são organizados em estruturas fixas, enquanto dados não estruturados não possuem uma organização predefinida e podem assumir várias formas, como texto, áudio, vídeo, imagens, entre outros.

# 1.4 Fontes de dados

A Ciência de Dados é uma disciplina interdisciplinar que se concentra na coleta, análise e interpretação de dados para obter insights valiosos e orientar a tomada de decisões. A escolha das estruturas e fontes de dados desempenha um papel fundamental nesse processo, uma vez que determina a eficácia da análise e a qualidade dos resultados obtidos.

image.png

**JSON (JavaScript Object Notation):**
O JSON é um formato leve e de fácil leitura que é amplamente utilizado para armazenar e trocar dados estruturados. Ele usa uma sintaxe baseada em pares de chave-valor, onde os dados são organizados em objetos aninhados. Cada objeto pode conter atributos (chaves) e seus valores correspondentes, que podem ser strings, números, booleanos, arrays, outros objetos JSON ou valores nulos. O JSON é particularmente útil para representar dados semiestruturados, como respostas de APIs da web, configurações de aplicativos e registros de eventos.

**CSV (Comma-Separated Values):**
O formato CSV é uma maneira simples de armazenar dados tabulares, onde cada linha representa um registro e os valores são separados por vírgulas (ou outro delimitador). Ele é amplamente utilizado para importar e exportar dados entre aplicativos e sistemas. Apesar de sua simplicidade, o CSV não oferece suporte direto para tipos de dados complexos ou estruturas aninhadas, limitando sua capacidade de representar informações mais ricas.

**Bancos de Dados Relacionais:**
Os bancos de dados relacionais são estruturas organizadas que usam tabelas para armazenar dados e relacionamentos entre eles. Eles usam linguagens como SQL (Structured Query Language) para consultar e manipular dados. Exemplos de bancos de dados relacionais incluem MySQL, PostgreSQL e Oracle. Esses bancos de dados são adequados para armazenar grandes volumes de dados estruturados e oferecem suporte a consultas complexas.

**Bancos de Dados NoSQL:**
Os bancos de dados NoSQL (Not Only SQL) são projetados para lidar com tipos de dados não estruturados ou semiestruturados. Eles são mais flexíveis do que os bancos de dados relacionais e incluem diversos modelos, como documentos, colunas, grafos e chave-valor. Exemplos populares são MongoDB (documentos), Cassandra (colunas) e Neo4j (grafos). Bancos de dados NoSQL são frequentemente usados em aplicações web e móveis que requerem escalabilidade e agilidade.

**Fontes de Dados da Web:**
A web é uma fonte rica de dados para a ciência de dados. APIs (*Application Programming Interfaces*) fornecem maneiras estruturadas de acessar dados de plataformas online, como redes sociais, serviços de mapeamento e previsão do tempo. Esses dados podem estar em formato JSON ou XML. A *Webscrapping* é outra técnica usada para extrair informações de páginas da web, transformando dados não estruturados em formatos utilizáveis.

# <h1> 1.5 Formatos de dados: CSV.</h1>


Quando os dados começaram a ser armazenados em discos rígidos e discos, o espaço era muito limitado. Para armazenar dados da forma mais eficiente possível, foram adotados padrões universais para os símbolos que seriam considerados dados válidos. Esses padrões, que ainda são usados atualmente, são chamados de ASCII:

ASCII - (pronuncia-se "As-Key") Código Padrão Americano para Intercâmbio de Informações:

128 caracteres são considerados "legais" em arquivos de dados.






*   Os arquivos ASCII são arquivos de texto simples. Eles podem ser confusos e não estar imediatamente prontos para uso em análises, mas temos métodos para carregar e limpar dados ASCII.

*   Os pontos de dados geralmente são delimitados por vírgulas, espaços ou tabulações, ou podem exigir um dicionário de dados para serem lidos.


*   Os arquivos ASCII são projetados para serem tão pequenos e universalmente portáteis quanto possível.

O formato ASCII mais comum é o CSV (comma-separated values, valores separados por vírgulas), no qual os pontos de dados individuais são separados por vírgulas. Um arquivo CSV é um arquivo de texto simples que pode ser aberto com o Notepad no Windows ou com o TextEdit no Mac. Cada linha dos dados aparece em uma nova linha do arquivo de texto, mas as colunas não necessariamente se alinham; em vez disso, as vírgulas representam novas colunas, de modo que a quarta e a quinta vírgulas em uma linha cercam a quinta coluna dos dados, por exemplo. Se os nomes das colunas estiverem incluídos nos dados, eles serão listados na primeira linha. Se você abrir um arquivo CSV em um editor de texto simples, ele terá a seguinte aparência:

## <h1> Observando o Dataframe para ver se ele foi carregado corretamente </h1>


Um dos maiores erros que alguém pode cometer ao carregar dados no Python (ou em qualquer ambiente de programação) é presumir que os dados foram carregados corretamente quando o código não apresenta um erro. Há muitas coisas que ainda podem dar errado. Os dados podem não ter lido corretamente os nomes das colunas, as dimensões podem estar erradas, os dados podem ter sido lidos de uma forma que fez com que os dados válidos fossem substituídos por valores ausentes, e assim por diante.

Há várias funções que são úteis para ver os dados e entender se eles foram carregados corretamente. A melhor estratégia é carregar os dados, examinar os dados carregados para entender quaisquer problemas e, se necessário, recarregar os dados para corrigir esses problemas desde o início.

A melhor maneira de visualizar o dataframe é digitar o nome do quadro de dados em uma célula de um notebook Jupyter. A saída da célula será renderizada como uma planilha e, nesse formato, ficará claro quais problemas, se houver, existem com o Dataframe. Um problema: para dataframes grandes (muitas linhas, muitas colunas ou ambos), a exibição de toda a planilha ocupa desnecessariamente muito espaço. Antes de exibir uma planilha, podemos definir os padrões para o número de linhas e colunas a serem exibidas da seguinte forma:

```python
pd.set_option('display.max_columns', 10) #display a maximumn of 10 columns
pd.set_option('display.max_rows', 10) #display a maximumn of 10 rows
```

Agora, novamente:

## <h1> Exibição do tipo de dados de cada coluna </h1>


Todos os dados armazenados na memória do Python recebem um tipo que controla como várias funções operam por padrão nesses dados. Por exemplo, o método .`describe()` fornece médias e outras estatísticas resumidas para dados numéricos e frequências para dados de texto. Se você pretende executar operações matemáticas em uma coluna de um quadro de dados, como incluir a coluna em um modelo de regressão, é importante primeiro confirmar que a coluna é lida com um tipo de dados apropriado. Os tipos de dados mais comuns são:

* **int64**: números inteiros com até 64 dígitos

* **float64**: números com decimais com até 64 dígitos no total

* **object**: texto ou dados numéricos codificados como texto

* **bool**: somente Verdadeiro ou Falso



Podemos usar o atributo `.dtypes` para exibir os tipos de dados de cada coluna. Como essas informações são armazenadas em uma tabela, precisamos desativar o limite padrão do número de linhas exibidas para ver todos os tipos de dados das colunas:

## <h1> Identificando e resolvendo problemas com dados e arquivos baseados em texto </h1>

Há muitas maneiras de organizar os arquivos de dados baseados em texto que causam problemas no carregamento dos dados no Python e no uso dos dados para análises. Observe que nem todos os problemas precisam ser resolvidos na etapa de carregamento: toda a biblioteca pandas existe para limpar os dados depois que eles são carregados na memória do Python. Mas alguns problemas são resolvidos de forma mais eficiente à medida que carregamos o arquivo. Aqui discutiremos alguns problemas comuns e estratégias para resolvê-los durante o carregamento dos dados.

**1) Comentários antes do cabeçalho**

```python
url = "https://raw.githubusercontent.com/jkropko/DS-6001/master/localdata/anes_example_toplines.csv"
import requests
file = requests.get(url)
print(file.text[0:1000])
```

Se você tentar carregar o arquivo de dados como ele existe atualmente, digitando:

```python
anes = pd.read_csv(url)
```

Gera o erro:

```python
ParserError: Error tokenizing data. C error: Expected 1 fields in line 5, saw 168
```

Esse erro ocorre porque a primeira linha informa à função `pd.read_csv()` que há uma coluna nos dados, pois não há vírgulas nessa linha. Portanto, no momento em que os nomes das colunas aparecem, há mais vírgulas do que o analisador sabe como lidar.

Para resolver esse problema, use o parâmetro header para informar ao `pd.read_csv()` quantas linhas do arquivo de dados são ocupadas pelo cabeçalho indesejado. Essa função sempre supõe que os nomes das colunas são a primeira linha dos dados após o cabeçalho.

```python
anes = pd.read_csv(url, header = 4).head(10)
anes
```

**2) Dados com códigos numéricos ausentes**

Esta versão usa -999 para representar vários valores ausentes, como quando um respondente da pesquisa se recusa a responder a uma pergunta:

```python
url = "https://raw.githubusercontent.com/jkropko/DS-6001/master/localdata/anes_example_missing.csv"
anes = pd.read_csv(url)
anes
```

Sempre podemos continuar com o quadro de dados como está e usar várias funções no pandas para substituir esses valores -999. Mas, nesse caso, é mais fácil ler esses valores como ausentes quando carregamos os dados. Podemos fazer isso com o parâmetro `na_values`. Então, todos os valores -999 são substituídos pelo caractere `NaN`, que o Python reconhece como um ponto de dados ausente.

```python
anes = pd.read_csv(url, na_values = -999)
anes
```

**3) Comentários dentro dos dados**

Suponha que você trabalhe com um arquivo de dados no qual o provedor de dados deixou comentários ao longo dos dados. Isso é uma heresia. Mas isso definitivamente pode acontecer. Esperamos que o comentarista tenha pensado o suficiente para colocar um caractere uniforme na frente desses comentários. Se for o caso, podemos remover os comentários durante o carregamento dos dados. Considere o seguinte arquivo de dados:

```python
url = "https://raw.githubusercontent.com/jkropko/DS-6001/master/localdata/anes_example_comments.txt"
anes = pd.read_csv(url, header = 4).head(10)
anes
```

Para ignorar esses comentários, use o parâmetro `comment`:

```python
anes = pd.read_csv(url, header = 4, comment = '@').head(10)
anes
```

**4) Dados sem nomes de colunas**

A função `pd.read_csv()` assume que os nomes das colunas estão listados na primeira linha do arquivo de dados, a menos que o parâmetro header (cabeçalho) seja usado, no qual ele assume que os nomes das colunas estão na primeira linha após o cabeçalho. Mas se os dados não contiverem uma linha de nomes de coluna, o Python usará a primeira linha de dados válidos como nomes de coluna. Criei uma versão dos dados do ANES com os nomes das colunas excluídos. Veja o que acontece quando carrego esse arquivo de dados normalmente:

```python
url = "https://raw.githubusercontent.com/jkropko/DS-6001/master/localdata/anes_example_nocolnames.csv"
anes = pd.read_csv(url).head(10)
anes
```

Os nomes das colunas não são descritivos do que elas representam. Mas pior do que isso, esses nomes são os pontos de dados da primeira linha dos dados. Isso significa que perdemos uma observação. Para lidar com esse problema, use o parâmetro `header=None`.

```python
anes = pd.read_csv(url, header=None).head(10)
anes
```

Embora ainda tenhamos o problema de os nomes das colunas não serem descritivos, eles são rotulados logicamente em ordem numérica da esquerda para a direita, e não perdemos a primeira observação. Se quisermos adicionar o prefixo "X" a cada um desses números, podemos adicionar o parâmetro `prefix="X"`:

# <h1> 1.6 Formatos de dados: JSON.</h1>


JSON significa JavaScript Object Notation (Notação de Objeto JavaScript) e é um dos formatos de dados mais usados para transferência de dados pela Internet. Se você estiver usando um sistema de transferência de dados baseado na Web, como uma API (mais sobre isso no próximo módulo), estará lidando com dados no formato JSON.

O formato de dados JSON é atribuído a Douglas Crockford, um arquiteto de Javascript do Yahoo! "Eu descobri o JSON. Não afirmo que inventei o JSON, porque ele já existia na natureza. O que eu fiz foi encontrá-lo, nomeá-lo e descrever como ele era útil".

Crockford e seus colaboradores desenvolveram o JSON em 2001 com o objetivo de criar uma linguagem para armazenar, exibir e transferir dados de uma forma que funcione em todos os navegadores. Em outras palavras, o objetivo do JSON é ser universal, de modo que os dados possam ser lidos independentemente do navegador ou do ambiente. O JSON também foi projetado para ser o mais leve possível, ou seja, o código JSON é o mais compacto possível.

O JSON sacrifica os recursos que expandem a funcionalidade da linguagem, como comentários, instruções de processamento ou atributos, para usar a quantidade mínima de código. A construção leve do JSON torna o código JSON muito mais rápido do que linguagens alternativas de intercâmbio de dados, como XML. Ela também ajuda a atingir a meta de universalidade: uma linguagem leve funciona dentro das infraestruturas existentes sem nenhum software adicional ou código estranho, porque toda linguagem de programação entende os dados de alguma forma, e uma estrutura de dados minimalista funciona com a ideia de cada linguagem sobre o que os dados devem ser.

A maior diferença entre o JSON (e outros formatos de intercâmbio de dados, como XML, PHP e YAML) e o CSV é que o JSON acomoda estruturas de dados baseadas em árvores, enquanto o CSV só lida com dados tabulares. Muitos aplicativos do mundo real exigem uma estrutura de dados baseada em árvore. Por exemplo, aqui estão os dois primeiros registros de dados falsos dos registros de clientes de uma empresa:

## <h1> Listas, Conjuntos e Dicionários </h1>


Primeiro, observe a chave quadrada de abertura `[`. Esse caractere diz ao Python para ler todos os registros seguintes como elementos de uma lista. Em uma lista Python, a ordem dos elementos é importante e os elementos podem ser repetidos, mas não recebem nomes. O primeiro elemento é denotado com o índice 0, o segundo elemento é denotado com o índice 1 e assim por diante. Por exemplo:

```python
my_list = [5,8,-9]
my_list[0]
```

Na sintaxe JSON, cada elemento individual da lista é um conjunto, denotado por chaves `{` e `}`. Um conjunto Python difere de uma lista Python porque a ordem dos elementos não importa (ele classifica os elementos automaticamente), não permite repetição e permite que os elementos sejam nomeados. Por exemplo, no código a seguir, observe como os 5s repetidos são removidos e como os elementos são classificados do menor para o maior:

```python
my_set = {5,5,5,8,-9}
my_set
```

Só é possível usar a chamada de elementos individuais de um conjunto se esses elementos tiverem nomes distintos:

```python
my_set = {'larry':5, 'curly':8, 'moe':-9}
my_set['larry']
```

Em Python, os conjuntos nos quais os elementos de um conjunto têm nomes são chamados de dicionários. Os nomes são chamados de chaves, e os próprios elementos são chamados de valores. Portanto, nesse caso, um elemento do dicionário tem a chave `"larry"` e o valor 5. Assim como os conjuntos, os dicionários não impõem repetição, mas aplicam essa restrição somente às chaves: várias chaves podem ter o mesmo valor, mas cada valor deve ter uma chave distinta. Neste exemplo, defino a chave `"larry"` duas vezes, e somente o último `"larry"` inserido na definição do dicionário é salvo:

```python
my_set = {'larry':15, 'larry':10, 'curly':10, 'moe':-9}
my_set
```

Isso significa que a notação JSON é o mesmo que uma lista em que cada elemento é um dicionário em Python. Cada dicionário representa um registro nos dados e, quando convertermos os dados em um dataframe tabular, cada registro ocupará uma linha nos dados.

## <h1> Carregamento e leitura de dados JSON em Python </h1>


O JSON, assim como o CSV e o ASCII, é um sistema baseado em texto para armazenamento de dados. Ao carregar dados JSON pela primeira vez no Python, os dados serão lidos como um bloco gigante de texto. Primeiro, precisamos fazer com que o Python entenda que o texto é, na verdade, um dado JSON organizado usando a biblioteca json. Quando pudermos trabalhar com os dados como JSON, poderemos pesquisar no caminho do índice dos dados JSON para extrair pontos de dados específicos. Podemos até mesmo construir loops para extrair listas de valores de todos os registros.

Há uma função importante da biblioteca requests e duas funções importantes na biblioteca json que devemos conhecer ao trabalhar com dados JSON:

* `json.loads()` converte o texto em um objeto que o Python reconhece como dados formatados em JSON.

* `json.dumps()` converte dados JSON em um bloco de texto.

## <h1> Pesquisando ao longo do caminho do índice JSON </h1>


Depois de usar json.loads(), o Python entende que os dados JSON são uma lista de listas. Agora podemos usar índices para extrair pontos de dados específicos. Os registros são números, começando com 0, e as chaves dentro de registros específicos podem ser chamadas pelo nome.

Por exemplo, para extrair o endereço de e-mail do quinto cliente (elemento 4), nos dados, eu digito:

```python
users_json[4]['email']
```

Se chamarmos campos que contenham vários recursos aninhados, o Python retornará um dicionário:

```python
users_json[0]['address']
```

Para extrair elementos que estão em vários níveis de aninhamento na árvore JSON, especifique as chaves na ordem que levam ao elemento desejado. Por exemplo, para extrair a coordenada latitudinal, navegamos para as chaves `address, geo e lat`:

```python
users_text = json.dumps(users_json)
type(users_text)
```

## <h1> Looping entre registros para extrair dados </h1>


Embora o JSON seja um formato de dados bem organizado e flexível que preserva muito do contexto significativo dos dados, a maioria das ferramentas analíticas e dos modelos estatísticos só pode operar com dados tabulares. Portanto, um grande desafio com os dados JSON é converter os dados em um formato tabular para serem salvos em um quadro de dados. Um método é construir um loop entre os registros para extrair os elementos desejados.

Antes de discutirmos esse método, não recomendo usar essa abordagem em geral. Os loops geralmente são lentos, e é muito mais rápido executar esses tipos de operações por meio da vetorização do código. A vetorização refere-se à capacidade de uma linguagem de programação de operar em um vetor inteiro de dados e trabalhar em vários elementos simultaneamente, em vez de trabalhar em cada elemento, um de cada vez. A vetorização é a especialidade da biblioteca numpy, e a maioria das funções do pandas também é vetorizada. Sugiro usar `pd.read_json()` e `pd.json_normalize()` para converter dados JSON em um quadro de dados, pois eles serão várias ordens de magnitude mais rápidos para converter todo o JSON em um quadro de dados. O looping só faz sentido quando precisamos extrair um pequeno número de recursos de dados JSON com um grande conjunto de recursos, porque com o loop chamamos diretamente os recursos que queremos e podemos evitar ter de trabalhar com recursos de que não precisamos.

Para iniciar o loop, escolha um índice para representar um registro. Para exibir todos os endereços de e-mail, digite:

```python
for u in users_json:
    print(u['email'])
```

Como alternativa, para colocar todos os endereços de e-mail em uma lista, você pode fazer um loop dentro de uma lista como esta:

```python
emails = [u['email'] for u in users_json]
emails
```

O truque é extrair os dados e salvá-los em um Dataframe ao mesmo tempo. Podemos fazer isso usando a função `pd.DataFrame()` e usando loops de lista dentro dessa função:

```python
users_df = pd.DataFrame(
    [u['name'], u['email'], u['company']['name']] for u in users_json
)
users_df.columns = ['name', 'email', 'company_name']
users_df
```

Observe que há muitas maneiras de construir um loop para extrair elementos JSON em um dataframe. Se você procurar no Stack Overflow, por exemplo, verá muitas abordagens diferentes, e isso pode ser confuso. Encontre uma abordagem que você entenda e se sinta confortável em usar, e siga-a. Não há muita diferença entre um loop e outro, pois o verdadeiro aprimoramento vem da vetorização.

## <h1> Armazenando dados de um JSON em um Dataframe </h1>


Se não houver muitos recursos nos dados, ou se você quiser manter todos os recursos nos dados de qualquer forma, a melhor e mais rápida maneira de converter dados JSON em um Dataframe tabular dependerá do fato se os dados JSON terem ou não uma estrutura aninhada.

## <h1> Situação 1: sem dados aninhados </h1>


Se o JSON não tiver aninhamento, ou seja, se cada chave estiver associada a um único ponto de dados e nenhuma chave estiver associada a um dicionário que contenha recursos adicionais, e também não incluir metadados, use as etapas a seguir:



1.   Use `requests.get()` para baixar os dados JSON brutos (a menos que você tenha outra forma de adquirir os dados brutos);
2.   Use `pd.read_json()` no atributo .text da saída de `requests.get()`;





O resultado será um dataframe no qual as colunas têm os mesmos nomes das chaves no arquivo JSON. Por exemplo, o JSON Placeholder tem um exemplo de conjunto de dados JSON que contém publicações latinas aleatórias em um blog. Esse JSON não contém metadados nem aninhamento. A melhor maneira de adquirir esse conjunto de dados e convertê-lo em um dataframe é a seguinte:

```python
posts = requests.get("https://jsonplaceholder.typicode.com/posts")
posts_df = pd.read_json(posts.text)
posts_df
```

## <h1> Situação 2: dados aninhados </h1>


Se o arquivo JSON contiver aninhamento a melhor estratégia é



1.   Usar `requests.get()` para baixar os dados JSON brutos (a menos que você tenha outra forma de adquirir os dados brutos)
2.   Usar `json.loads()` no atributo `.text` da saída da etapa 1 para registrar os dados como uma lista no Python
3. Use a função `pd.json_normalize()` na lista que é a saída da etapa 2



A função `pd.json_normalize()` armazena cada característica nos dados em uma coluna separada, independentemente do número de níveis de aninhamento que deve ser analisado para encontrar a característica.

Cada coluna tem o mesmo nome da chave da qual foi extraído o recurso. Para recursos que estão aninhados em outros recursos, o `pd.json_normalize()` usa cada chave no caminho para o ponto de dados para construir o nome da coluna, separado por pontos. Por exemplo, os dados dos usuários com os quais trabalhamos acima contêm até três níveis de aninhamento. Portanto, os dados lat são armazenados em uma coluna chamada `address.geo.lat`, já que tivemos que navegar até `"address"`, depois `"geo"` e depois `"lat"` no JSON para encontrar esses dados:

```python
users = requests.get("https://jsonplaceholder.typicode.com/users")
users_json = json.loads(users.text)
users_df = pd.json_normalize(users_json)
users_df
```

```python
from math import radians, sin, cos, sqrt, atan2

def haversine_distance(lat1, lon1, lat2, lon2):
    R = 6371.0  # Raio da Terra em quilômetros
    dlat = radians(lat2 - lat1)
    dlon = radians(lon2 - lon1)
    a = sin(dlat / 2)**2 + cos(radians(lat1)) * cos(radians(lat2)) * sin(dlon / 2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))
    distance = R * c
    return distance

# Coordenadas de um local de referência
reference_lat = 36.7
reference_lon = 71.6
```
