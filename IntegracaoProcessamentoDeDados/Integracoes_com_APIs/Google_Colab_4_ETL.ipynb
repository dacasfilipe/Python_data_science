#1.2  Conceito de ETL e seu papel na preparação de dados para análise..






O mundo atual é movido por dados. De cada transação financeira a cada clique em um site, vastas quantidades de informações são geradas diariamente. No entanto, antes que esses dados possam se transformar em insights significativos, eles frequentemente precisam passar por um processo fundamental: Extração, Transformação e Carga, ou ETL (*Extract, Transform and Load*).

O ETL representa um dos alicerces cruciais da análise de dados moderna, servindo como uma ponte entre os dados brutos e a compreensão informada. O processo é comparável a um refinamento cuidadoso, onde matérias-primas ásperas e desorganizadas são lapidadas até revelarem seu verdadeiro valor. Nesse contexto, o ETL desempenha um papel central na preparação desses dados brutos para análises detalhadas e interpretações significativas.

O conceito de ETL gira em torno de três fases interligadas:

**1. Extração:**
A primeira etapa envolve a coleta de dados de várias fontes. Essas fontes podem variar desde bancos de dados, planilhas, arquivos CSV até APIs da web e fontes de dados em tempo real. A extração visa reunir os dados brutos em um local centralizado, pronto para serem processados.

**2. Transformação:**
Uma vez extraídos, os dados passam por uma série de transformações. Essa fase é crucial, pois é aqui que os dados são limpos, reformatados e refinados. A limpeza envolve a remoção de ruídos, tratamento de valores ausentes e correção de inconsistências. Além disso, a transformação inclui a conversão de tipos de dados, agregações, normalizações e outras operações que moldam os dados para se tornarem mais coerentes e úteis para análise.

**3. Carga:**
Após serem transformados, os dados são carregados em um local de destino, geralmente um armazenamento de dados ou um banco de dados preparado para análises posteriores. Esse local deve ser projetado para suportar consultas eficientes e permitir o acesso rápido aos dados processados.

O papel do ETL na preparação de dados para análise é fundamental. Ele garante que os dados brutos, frequentemente desorganizados e sujeitos a erros, sejam refinados e organizados de maneira apropriada. Isso, por sua vez, facilita a obtenção de insights precisos e a tomada de decisões fundamentadas.

image.png

Os benefícios do ETL se estendem a várias áreas:

* **Qualidade dos Dados**: Através das etapas de limpeza e transformação, o ETL ajuda a garantir a qualidade dos dados, reduzindo erros e inconsistências que poderiam levar a análises equivocadas.

* **Consistência:** A transformação dos dados garante que eles estejam consistentes em termos de formato, tipos e valores, permitindo comparações e análises confiáveis.

* **Eficiência Analítica:** Ao preparar dados de maneira estruturada, o ETL permite que analistas e cientistas de dados concentrem-se nas análises em vez de gastar tempo na preparação inicial.

* **Escalabilidade:** À medida que os conjuntos de dados crescem, o ETL pode ser dimensionado para lidar com volumes maiores, garantindo que a preparação dos dados continue sendo eficaz.

# 1.3 Processos de extração, transformação e carga em detalhes.




## 1.3.1 Extração (*Extraction*)

A primeira etapa do processo ETL é a Extração, que envolve a coleta de dados brutos de várias fontes diferentes. Essa etapa é crucial para obter as informações necessárias para análises e tomada de decisões informadas. Vamos explorar em detalhes o processo de Extração no contexto do ETL, acompanhado de exemplos para ilustrar cada etapa.

**1. Identificação de Fontes de Dados:**
A extração começa com a identificação das fontes de dados relevantes para o processo. Isso pode incluir bancos de dados internos, planilhas, arquivos CSV, fontes externas via APIs e feeds em tempo real.

**Exemplo:** Se você está analisando dados de vendas, suas fontes podem incluir bancos de dados de vendas, sistemas de pagamento e plataformas de e-commerce.

**2. Acesso aos Dados:**
Uma vez identificadas as fontes, é necessário acessar os dados. Isso pode envolver o uso de consultas SQL, APIs de acesso a dados ou a leitura direta de arquivos.

**Exemplo:** Se os dados de vendas estão armazenados em um banco de dados, você pode usar consultas SQL para recuperar as informações relevantes.

**3. Extração de Dados Brutos:**
Nesta etapa, os dados brutos são extraídos das fontes identificadas. Isso pode envolver a recuperação de linhas de tabelas, registros de logs ou qualquer outra forma de dados que precisa ser processada.

**Exemplo:** Ao extrair dados de vendas, você pode obter informações como data da venda, ID do cliente, valor da transação e produtos comprados.

**4. Captura de Dados Incrementais ou Completa:**
Você pode optar por extrair apenas os novos dados desde a última extração (captura incremental) ou extrair todos os dados disponíveis (captura completa), dependendo da frequência das atualizações e do volume de dados.

**Exemplo:** Se você está extraindo dados de um sistema de logs, a captura incremental pode envolver apenas os novos registros gerados desde a última extração.

**5. Tratamento de Dados:**
Durante a extração, é importante considerar o tratamento inicial dos dados, como a conversão de tipos de dados, a remoção de dados irrelevantes ou a limpeza de valores ausentes.

**Exemplo:** Se você está coletando dados de clientes, pode precisar tratar campos como números de telefone para remover caracteres especiais e garantir consistência.

**6. Validação de Dados:**
Antes de passar para a próxima etapa, é crucial validar os dados extraídos para garantir sua integridade e precisão. Isso pode envolver verificações de formato, verificação de duplicatas ou outras regras de validação.

**Exemplo:** Verificar se os valores de datas estão no formato correto e se não há entradas duplicadas nos dados de vendas.

A etapa de Extração no ETL é a base para todo o processo, pois é a partir daqui que os dados brutos são adquiridos para serem transformados e posteriormente carregados em um formato adequado para análises. A qualidade da extração tem um impacto direto na qualidade dos insights que podem ser obtidos ao longo do processo ETL.

##1.3.2 Transformar (*Transform*)

A segunda etapa do ciclo ETL (Extração, Transformação e Carga) é a Transformação, que desempenha um papel crucial na preparação dos dados extraídos para análises e interpretações significativas. Durante essa fase, os dados brutos são limpos, reformatados e transformados em um formato coerente e útil para análise. Vamos explorar em detalhes o processo de Transformação no contexto do ETL, acompanhado de exemplos para ilustrar cada etapa.

**1. Limpeza de Dados:**
A limpeza envolve a identificação e a correção de erros e inconsistências nos dados brutos. Isso pode incluir a remoção de valores ausentes, a padronização de formatos e a correção de erros de digitação.

**Exemplo:** Se há registros de vendas com valores ausentes em campos críticos, como data da venda, a limpeza envolverá a decisão de remover esses registros ou preencher os valores ausentes de maneira apropriada.

**2. Conversão de Tipos de Dados:**
Durante a transformação, é comum que os tipos de dados sejam convertidos para garantir consistência e adequação às análises planejadas. Isso pode envolver a conversão de números, datas e outros tipos de dados.

**Exemplo:** Se os números de vendas estiverem representados como strings, você pode convertê-los para números inteiros ou decimais para cálculos mais precisos.

**3. Agregação e Sumarização:**
A agregação envolve a criação de resumos de dados, como médias, somas e contagens, para permitir análises de alto nível e identificação de tendências.

**Exemplo:** Ao analisar dados de vendas, você pode calcular a receita total, a média de vendas por cliente ou a quantidade média de produtos por transação.

**4. Transformações de Valores:**
Transformar valores envolve aplicar regras específicas para modificar ou derivar novos dados com base nos dados originais.

**Exemplo:** Se você tem dados de vendas com preços em diferentes moedas, pode criar uma nova coluna convertendo todos os preços para uma moeda comum.

**5. Normalização e Padronização:**
A normalização envolve a organização dos dados em um formato uniforme, o que é particularmente útil ao trabalhar com dados de diferentes origens.

**Exemplo:** Normalizar os nomes dos produtos para evitar duplicatas, como "Smartphone" e "Celular Inteligente".

**6. Mapeamento e Enriquecimento de Dados:**
Mapear dados envolve relacionar campos em diferentes conjuntos de dados para criar informações mais ricas e completas.

**Exemplo:** Adicionar informações geográficas a dados de vendas, relacionando os códigos de região aos nomes dos estados.

**7. Aplicação de Regras de Negócios:**
Durante a transformação, as regras de negócios podem ser aplicadas para segmentar dados, classificar registros ou derivar informações importantes.

**Exemplo:** Segmentar clientes em grupos com base em suas compras recorrentes para direcionar campanhas de marketing específicas.

A etapa de Transformação no ETL é onde os dados brutos são moldados e preparados para a análise subsequente. É aqui que a verdadeira mágica acontece, transformando dados desorganizados em informações valiosas, prontas para revelar padrões, insights e oportunidades ocultas. A qualidade e eficácia da transformação afetam diretamente a qualidade dos insights que podem ser obtidos nas fases subsequentes do ciclo ETL.

##1.3.3 Carga (*Load*)



O processo de Carga (Load), como parte do ciclo ETL (Extração, Transformação e Carga), é a etapa final e crucial para tornar os dados processados e transformados disponíveis para análises e consultas. Essa fase envolve a transferência dos dados preparados para um local de destino, geralmente um repositório de dados, armazém de dados ou banco de dados, onde podem ser acessados e consultados de maneira eficiente. Vamos explorar em detalhes o processo de Carga no contexto do ETL, acompanhado de exemplos para ilustrar cada etapa.

**1. Preparação do Armazenamento:**
Antes de carregar os dados, é essencial criar uma estrutura de armazenamento que seja adequada para as análises planejadas. Isso pode envolver a criação de tabelas, índices e definição de relacionamentos. A escolha do sistema de gerenciamento de banco de dados (DBMS) é crucial, pois ele deve suportar os requisitos de armazenamento e consulta.

**Exemplo:** Imagine que você está preparando dados de vendas de uma loja online. Nesse caso, você pode criar tabelas para armazenar informações sobre produtos, clientes, pedidos e transações.

**2. Correspondência de Campos e Transformações Finais:**
Antes de carregar os dados, é importante garantir que os campos mapeados na fase de Transformação correspondam às colunas nas tabelas de destino. Isso pode envolver conversões finais de tipos de dados e ajustes necessários.

**Exemplo:** Se você estiver carregando dados de datas de vendas, pode ser necessário converter formatos de datas para se adequar ao padrão do DBMS.

**3. Carregamento Incremental ou Completo:**
Você pode optar por carregar apenas os novos dados desde a última carga (carregamento incremental) ou carregar todos os dados novamente (carregamento completo). A escolha depende da frequência das atualizações e do volume de dados.

**Exemplo:** Se você está lidando com dados de vendas diárias, um carregamento incremental pode envolver apenas as vendas do dia atual.

**4. Carregamento Direto ou Staging:**
Existem duas abordagens principais para o carregamento: carregamento direto e carregamento para uma área de preparação chamada "staging". O carregamento direto envia os dados diretamente para as tabelas de destino, enquanto o carregamento para staging permite validações e verificações adicionais antes do carregamento final.

**Exemplo:** Antes de carregar dados de vendas para a tabela final, você pode carregá-los em uma área de staging para validar se todos os campos obrigatórios estão presentes.

**5. Manutenção de Integridade e Consistência:**
Durante o carregamento, é importante garantir a integridade e a consistência dos dados. Isso pode envolver a aplicação de restrições de chave primária, restrições de integridade referencial e validações de negócios.

**Exemplo:** Verificar se os IDs dos clientes correspondem aos IDs existentes na tabela de clientes antes de associar vendas a clientes específicos.

**6. Atualização de Metadados:**
Atualize os metadados relacionados aos dados carregados, como data e hora da última carga, status de carregamento e outras informações relevantes.

**Exemplo:** Registre a data e a hora em que os dados foram carregados pela última vez para monitoramento e auditoria.

O processo de Carga no ETL é a culminação de todo o trabalho realizado na Extração e Transformação, garantindo que os dados preparados estejam disponíveis para análises subsequentes. O objetivo é criar um ambiente em que os dados possam ser acessados de maneira eficiente, confiável e pronta para serem explorados para insights valiosos.

#1.4 Frameworks de ETL.


##1.4.1 O que é um *Framework*?

Um *framework* é a base sobre a qual os programas de software são criados. Ele serve de base para os desenvolvedores de software, permitindo que eles criem uma variedade de aplicativos para determinadas plataformas. É um conjunto de funções e classes predefinidas usadas para se conectar com o software do sistema e lidar com entradas e saídas.

Ele simplifica a vida de um desenvolvedor e, ao mesmo tempo, permite que ele use determinadas extensões e torna os aplicativos on-line escalonáveis e passíveis de manutenção.

##1.4.2 Frameworks de ETL Automatizados.


Existem várias ferramentas e frameworks automatizados de ETL (Extração, Transformação e Carga) disponíveis no mercado, que visam simplificar e agilizar o processo de preparação de dados para análises. Alguns dos frameworks mais populares incluem:

**Apache NiFi:** Uma ferramenta de fluxo de dados que permite a automação de ETL e a transferência de dados entre sistemas heterogêneos. Ele oferece uma interface gráfica para configurar fluxos de dados complexos e suporta diversas transformações e processamentos.

**Talend:** Um software de integração de dados de código aberto que oferece uma ampla gama de recursos para ETL, incluindo extração, transformação, carregamento, integração de dados e muito mais. Ele possui uma interface intuitiva para criar fluxos de trabalho de ETL complexos.

**Apache Spark:** Embora seja mais conhecido como um framework de processamento de dados em grande escala, o Spark também inclui módulos para processamento de ETL. O Spark oferece transformações poderosas e distribuídas que podem ser usadas para manipular dados antes de serem carregados.

**Apache Airflow:** Uma plataforma de agendamento e orquestração que também pode ser usada para automatizar fluxos de trabalho de ETL. Ele permite definir, agendar e monitorar processos de ETL complexos, garantindo que eles sejam executados de forma confiável.

**Microsoft SQL Server Integration Services (SSIS):** Uma ferramenta de ETL amplamente usada para ambientes Microsoft, que oferece uma interface gráfica para criar pacotes de ETL complexos. É especialmente útil para integração de dados em sistemas SQL Server.

**Informatica PowerCenter:** Uma ferramenta líder no mercado de ETL, que oferece uma variedade de recursos avançados para transformação e integração de dados. Ele é conhecido por sua escalabilidade e suporte a uma ampla variedade de fontes e destinos.

**Google Cloud Dataflow:** Uma solução de processamento de dados em tempo real e em lote oferecida pela Google Cloud. Pode ser usado para criar pipelines de ETL escalonáveis usando a programação Apache Beam.

**AWS Glue:** Um serviço gerenciado da Amazon Web Services que facilita a criação de pipelines de ETL. Ele oferece recursos como descoberta automática de esquema e escalabilidade para processar grandes volumes de dados.

**Pentaho Data Integration:** Uma solução open source que oferece recursos completos de ETL, incluindo extração, transformação, carregamento e agendamento de fluxos de trabalho.

**Matillion:** Uma ferramenta focada em ETL que foi projetada especificamente para integração de dados em ambientes de nuvem, como AWS e Snowflake.

Lembre-se de que a escolha do framework de ETL depende das necessidades específicas do seu projeto, da infraestrutura de tecnologia que você está usando e das características das fontes e destinos dos seus dados. Cada ferramenta possui suas próprias vantagens e desvantagens, e a escolha certa dependerá das suas metas e requisitos.

##1.5.1 Exemplo 1 - API Externa + ETL.

Este código ilustra um processo básico de ETL (Extração, Transformação e Carregamento) usando as bibliotecas `requests` e `pandas` em Python. Aqui está uma explicação resumida das partes principais do código:

**1. Importações de bibliotecas:** Importa as bibliotecas necessárias, `requests` para fazer solicitações HTTP à API e pandas para trabalhar com DataFrames.

**2. Passo 1: Extração `(extract_data)`:** Define uma função chamada `extract_data` que realiza a etapa de extração dos dados. Ele faz uma solicitação GET à URL "https://jsonplaceholder.typicode.com/posts", que é uma API de exemplo para testes. Ele recebe a resposta e converte os dados JSON em um formato Python.

**3. Passo 2: Transformação `(transform_data)`:** Define uma função chamada `transform_data` que realiza a etapa de transformação dos dados brutos. Ele percorre cada entrada nos dados brutos (postagens) e extrai informações específicas, como o título, o corpo e o ID do usuário. Em seguida, ele cria um dicionário com essas informações transformadas e os adiciona a uma lista.

**4. Passo 3: Carregamento `(load_data)`:** Define uma função chamada `load_data` que realiza a etapa de carregamento dos dados transformados em um DataFrame pandas. Ele converte a lista de dicionários transformados em um DataFrame.

**5. Função principal do ETL `(etl_process)`:** Esta é a função principal que executa o processo completo de ETL. Ela chama as funções de extração, transformação e carregamento sequencialmente para obter dados brutos, transformá-los em um formato adequado e carregá-los em um DataFrame.

```python
import requests
import pandas as pd

# Passo 1: Extração
def extract_data():
    url = "https://jsonplaceholder.typicode.com/posts"
    response = requests.get(url)
    data = response.json()
    return data

# Passo 2: Transformação
def transform_data(raw_data):
    transformed_data = []
    for entry in raw_data:
        transformed_entry = {
            'Title': entry['title'],
            'Body': entry['body'],
            'User ID': entry['userId']
        }
        transformed_data.append(transformed_entry)
    return transformed_data

# Passo 3: Carregamento
def load_data(transformed_data):
    df = pd.DataFrame(transformed_data)
    return df

# Função principal do ETL
def etl_process():
    # Passo 1: Extração
    raw_data = extract_data()

    # Passo 2: Transformação
    transformed_data = transform_data(raw_data)

    # Passo 3: Carregamento
    df = load_data(transformed_data)

    return df
```

```python
transformed_df = etl_process()
```

## 1.5.2 Exemplo 2 - API Externa (OpenWeatherMAP) + ETL.

Neste exemplo, usamos a API do OpenWeatherMap para extrair dados climáticos de uma cidade específica. A função extract_data faz uma solicitação HTTP à API com uma chave de API válida. A função transform_data extrai informações relevantes dos dados brutos e as organiza em um dicionário. A função load_data carrega os dados transformados em um DataFrame pandas.

Lembre-se de substituir **"YOUR_OPENWEATHERMAP_API_KEY"** pela sua chave de API do OpenWeatherMap. Este é apenas um exemplo básico, e em projetos reais, você precisaria considerar gerenciamento de erros, lidar com respostas assíncronas, implementar processamento paralelo para várias cidades, entre outros aspectos.

```python
import requests
import pandas as pd

# Passo 1: Extração
def extract_data(city):
    api_key = "SUA_API_KEY"
    base_url = f"http://api.openweathermap.org/data/2.5/weather?q={city}&appid={api_key}"
    response = requests.get(base_url)
    data = response.json()
    return data

# Passo 2: Transformação
def transform_data(raw_data):
    print(raw_data)
    transformed_data = {
        'City': raw_data['name'],
        'Country': raw_data['sys']['country'],
        'Temperature (C)': raw_data['main']['temp'] - 273.15,
        'Weather': raw_data['weather'][0]['description']
    }
    return transformed_data

# Passo 3: Carregamento
def load_data(transformed_data):
    df = pd.DataFrame([transformed_data])
    return df

# Função principal do ETL
def etl_process(city):
    # Passo 1: Extração
    raw_data = extract_data(city)

    # Passo 2: Transformação
    transformed_data = transform_data(raw_data)

    # Passo 3: Carregamento
    df = load_data(transformed_data)

    return df
```

```python
city_name = "London"  # Nome da cidade para obter os dados climáticos
transformed_df = etl_process(city_name)
```

A função `extract_data` faz uma solicitação HTTP à API, filtrando por linguagem e classificando por estrelas. A função `transform_data` extrai informações relevantes dos dados brutos e as organiza em uma lista de dicionários. A função `load_data` carrega os dados transformados em um DataFrame pandas. Note que, neste exemplo o processo de load, estamos fazendo o carregamento em arquivos CSV, EXCEL e JSON.

Lembre-se de que este exemplo considera apenas uma página de resultados. Em projetos reais, você precisaria lidar com a paginação de resultados para obter todos os dados, gerenciar possíveis erros de API, implementar autenticação (se necessário), entre outros aspectos.

```python
import requests
import pandas as pd

# Passo 1: Extração
def extract_data(language, page=1):
    url = f"https://api.github.com/search/repositories"
    params = {
        'q': f'language:{language}',
        'sort': 'stars',
        'order': 'desc',
        'page': page,
        'per_page': 100
    }
    headers = {
        'Accept': 'application/vnd.github.v3+json'
    }
    response = requests.get(url, params=params, headers=headers)
    data = response.json()
    return data

# Passo 2: Transformação
def transform_data(raw_data):
    transformed_data = []
    for item in raw_data['items']:
        transformed_entry = {
            'Repository Name': item['name'],
            'Owner': item['owner']['login'],
            'Stars': item['stargazers_count'],
            'Forks': item['forks_count'],
            'URL': item['html_url']
        }
        transformed_data.append(transformed_entry)
    return transformed_data

# Passo 3: Carregamento
def load_data(transformed_data, output_format='csv'):
    df = pd.DataFrame(transformed_data)

    if output_format == 'csv':
        df.to_csv('output_data.csv', index=False)
    elif output_format == 'excel':
        df.to_excel('output_data.xlsx', index=False)
    elif output_format == 'json':
        df.to_json('output_data.json', orient='records')

    return df

# Função principal do ETL
def etl_process(language, output_format='csv'):
    # Passo 1: Extração (página 1)
    raw_data = extract_data(language)

    # Passo 2: Transformação
    transformed_data = transform_data(raw_data)

    # Passo 3: Carregamento
    df = load_data(transformed_data, output_format)

    return df
```

```python
programming_language = "python"

# Saída para CSV
transformed_df_csv = etl_process(programming_language, output_format='csv')
print("CSV Data:")
print(transformed_df_csv.head())

# Saída para Excel
transformed_df_excel = etl_process(programming_language, output_format='excel')
print("Excel Data:")
print(transformed_df_excel.head())

# Saída para JSON
transformed_df_json = etl_process(programming_language, output_format='json')
print("JSON Data:")
print(transformed_df_json.head())
```

**3. Declaração de Base (Base):** Cria uma instância da classe `declarative_base()` do SQLAlchemy, que é a base para criar classes que mapeiam para tabelas do banco de dados.

**4. Definição da classe `RepositoryData`:** Cria uma classe que representa a estrutura da tabela `repository_data` no banco de dados. Essa classe herda da classe `Base`. Os atributos da classe correspondem às colunas da tabela.

**5. Função load_data(transformed_data):** Essa função carrega os dados de um DataFrame para a tabela `repository_data` no banco de dados MySQL.

```python
from sqlalchemy import create_engine
from sqlalchemy.orm import sessionmaker
from sqlalchemy.exc import SQLAlchemyError
from sqlalchemy.ext.declarative import declarative_base
from sqlalchemy import Column, Integer, String

# Configurações do banco de dados
db_config = {
    'host': 'your_host',
    'database': 'your_database',
    'user': 'your_user',
    'password': 'your_password'
}

Base = declarative_base()

# Definindo a estrutura da tabela
class RepositoryData(Base):
    __tablename__ = 'repository_data'
    id = Column(Integer, primary_key=True, autoincrement=True)
    repository_name = Column(String(255))
    owner = Column(String(255))
    stars = Column(Integer)
    forks = Column(Integer)
    url = Column(String(255))

# Passo 3: Carregamento
def load_data(transformed_data):
    df = pd.DataFrame(transformed_data)

    try:
        engine = create_engine(f"mysql+mysqlconnector://{db_config['user']}:{db_config['password']}@{db_config['host']}/{db_config['database']}")
        Session = sessionmaker(bind=engine)

        Base.metadata.create_all(engine)  # Crie a tabela se ela não existir

        session = Session()

        records_to_insert = []
        for _, row in df.iterrows():
            record = RepositoryData(repository_name=row['Repository Name'],
                                    owner=row['Owner'],
                                    stars=row['Stars'],
                                    forks=row['Forks'],
                                    url=row['URL'])
            records_to_insert.append(record)

        session.add_all(records_to_insert)
        session.commit()
        print("Dados inseridos no banco de dados.")
    except SQLAlchemyError as e:
        print("Erro ao conectar ou inserir dados:", e)
    finally:
        if session:
            session.close()
            print("Sessão fechada.")

    return df

# Função principal do ETL
def etl_process(language):
    # ... (mesmo código de extração e transformação)
    # Passo 3: Carregamento
    df = load_data(transformed_data)

    return df


programming_language = "python"
transformed_df = etl_process(programming_language)
print(transformed_df)
```
