# 1.2 APIs: Facilitando a Integração e a Comunicação

Uma API (Application Programming Interface) é um conjunto de regras, protocolos e ferramentas que permitem que diferentes softwares se comuniquem e interajam entre si. Ela serve como uma ponte entre sistemas, permitindo que aplicativos solicitem e compartilhem dados e funcionalidades de maneira padronizada. Em essência, as APIs facilitam a integração, ao oferecer uma interface clara e definida para a interação entre sistemas diversos.

Comumente, usa-se os termos "expõe" ou "implementa" para indicar que uma aplicação
permite acesso as suas funcionalidades e recursos de uma forma padronizada através do uso
de uma API.
O propósito de uma API consiste em simplificar o processo de integração entre aplicações, escondendo inúmeros detalhes de implementação e expondo apenas objetos e funções
necessários ao desenvolvimento.
Por mais simples que uma aplicação seja, ela fará uso de componentes expostos por outra
API. E, adicionalmente, uma API poderá fazer uso de componentes expostos por outras APIs.
Dessa forma, cada software acaba sendo usado como componentes em outro software, formando uma cascada de integração.

# 1.3 API REST (*Representational State Transfer*).

Uma API de Transferência de Estado Representacional (REST), ou Representational State
Transfer do inglês, trata-se de um estilo de arquitetura voltado para construção de serviços e
micro-serviços web que adotam um conjunto de princípios e práticas específicos.
Esse conjunto de princípios e práticas estabelece um contrato entre consumidor (cliente)
e produtor (servidor) de informações, ou seja, formato do conteúdo requirido pelo consumidor
(requisição) e o formato do conteúdo requirido pelo produtor (resposta).

Diferentemente de outros protocolos, REST permite inter-operabilidade entre aplicações
de diferentes plataformas e escritas em diferentes linguagens de programação, bem como, a
adoção de diferentes formatos de dados, sendo JSON o mais amplamente usado.
Para que uma API seja considerada uma API REST é necessário que a aplicação siga os
seguintes princípios arquiteturais:



*  **Interface uniforme:** Define que o servidor deve transferir as informações em um formato
padrão independente da origem da requisição. Bem como, o servidor ser capaz de retornar
todos os dados associados a um recurso necessários e meta-dados sobre como o cliente
poderá processar o recurso;
*   **Desacoplamento entre cliente e servidor:** Define que servidor e cliente devem ser totalmente independentes um do outro, sendo que a única informação necessária ao cliente é o
URL do recurso desejado;

*   **Ausência de estado:** Define que o servidor deve processar as requisições de um cliente
independentemente das requisições anteriores. Sendo que cada requisição deve conter todos
os dados necessários para seu processamento e deve ser isolada do processamento das
demais requisições;

*   **Sistema em camadas:** Define que a comunicação entre as partes envolvidas seja transparente em relação as diferentes camadas intermediárias, onde um requisição pode ser recebida
por um servidor e encaminhada para outro servidor usando uma nova API REST;

*   **Capacidade de armazenamento temporário:** Define que tanto servidor como intermediário
e cliente devem ser capazes de armazenar cache temporário a fim de melhorar o tempo de
resposta do servidor;

*   **Código sob demanda:** Define que opcionalmente o servidor deve ser capaz de transferir
código para processamento no ciente a fim de estender ou personalizar temporariamente
alguma funcionalidade no cliente.


Sendo que os principais benefícios desse tipo de API envolvem escalabilidade, flexibilidade
e independência. A escalabilidade é alcançada principalmente pela ausência de estado entre
cliente e servidor porque o servidor não precisa reter informações a respeito dos clientes e,
também, pela capacidade de armazenamento temporário (cache) porque promove melhoria
geral de desempenho no processamento das requisições. Enquanto que a separação total entre
cliente e servidor permite maior flexibilidade de toda a arquitetura envolvida porque possibilidade que cada parte possa evoluir independentemente e separadamente, desde que o contrato
estabelecido entre cliente e servidor seja respeitado. E, por fim, a liberdade de escolha de
plataforma e linguagem de programação garante independência das tecnologias usadas pelo
cliente e servidor, pois possibilidade que aplicações cliente sejam desenvolvida para plataforma
totalmente diferentes da adotada pelo servidor e, também, usando linguagens de programação
distintas.


# 1.4 Como funciona uma API Rest.

O funcionamento básico de uma API REST é semelhante ao processo de navegação na
internet realizada por um navegador web.

image.png

Primeiramente, a aplicação (cliente) envia uma solicitação, também chamada de requisição,
a API REST (servidor) para ter acesso a algum recurso ou lista de recursos. Em seguida, a
API REST realiza o processamento da solicitação fazendo uso de diversos outros componentes,
tais como repositório de arquivos local, repositório de arquivos em nuvem, banco de dados, ou
até mesmo outras APIs REST. Finalmente, a API REST retorna o resultado do processamento,
também chamado de resposta, em um formato de dados padronizado, tais como JSON ou XML.
Também, é possível realizar uma solicitação para ter acesso a documentação da API REST
e catalogar todas os recursos e ações disponíveis para manipulação desses recursos. Embora
não exista uma padronização oficial para documentação de uma API REST, é bastante comum
realizar essa documentação com uso do padrão Swagger.

# 1.4 Requisições e respostas HTTP.

Todas as APIs que usam o protocolo HTTP como formato de transporte para solicitações e respostas são consideradas "serviços da Web". Com os serviços da Web, o cliente que faz a solicitação do recurso e o servidor de API que fornece a resposta podem usar qualquer linguagem de programação ou plataforma - não importa, pois a solicitação e a resposta da mensagem são feitas por meio de um protocolo da Web HTTP comum.

O protocolo da Web faz parte da beleza dos serviços da Web: eles são independentes de linguagem e, portanto, interoperáveis em diferentes plataformas e sistemas. Ao documentar uma API REST, não importa se os engenheiros criam a API com Java, Ruby, Python ou alguma outra linguagem. As solicitações são feitas por HTTP e as respostas são retornadas por HTTP.

## 1.4.1 Requisições HTTP.

Uma solicitação ou requisição REST precisa conter os seguintes componentes:
* **Identificador de recurso exclusivo:** Cada recurso disponível no servidor é identificado unicamente através de um URL. Esse URL é utilizado pelo cliente a fim de indicar qual o recurso
desejado;
* **Método:** Método informa ao servidor o que precisa ser feito com o recurso indicado pelo
cliente. Sendo que os métodos mais comuns são: POST, GET, PUT e DELETE, correspondendo
às ações de criação, leitura, modificação e remoção, ou create, read, update e delete (CRUD)
do inglês;

 * POST: Informa ao servidor que o cliente deseja persistir os dados de um novo recurso no
URL especificado.
 * GET: Informa ao servidor que o cliente deseja acessar recursos localizados no URL especificado.
 * PUT: Informa ao servidor que o cliente deseja atualizar os dados de um recurso existente
no URL especificado.
 * DELETE: Informa ao servidor que o cliente deseja remover recursos localizados no URL
especificado.

* **Corpo da requisição:** Alguns métodos necessitam de dados adicionais, enviados no corpo
da requisição, para serem processados com êxito, como: POST e PUT;
* **Parâmetros:** Algumas requisições podem conter parâmetros adicionais que fornecem ao
servidor mais detalhes sobre como o processamento da requisição deve ser realizada. Esses
parâmetros podem ser enviados de diferentes maneiras pelo cliente, tais como: parte do
caminho ou parâmetros de consulta do URL, cookies adicionados ao cabeçalho da requisição
ou itens adicionais adicionados ao cabeçalho da requisição.


```python
###Método POST

# Importando a biblioteca 'requests' para fazer requisições HTTP.
import requests
# Importando o módulo 'json' para lidar com dados em formato JSON.
import json
# URL da API que será usada para criar um novo usuário.
url = 'https://jsonplaceholder.typicode.com/users'

# Dados do novo usuário que serão enviados para a API.
new_user = {
    'name': 'Novo Usuário',
    'username': 'novousuario',
    'email': 'novousuario@example.com'
}

# Fazendo uma requisição POST para a URL da API, enviando os dados do novo usuário em formato JSON.
response = requests.post(url, json=new_user)

# Imprimindo o código de status da resposta HTTP.
print(response.status_code)
# A função 'json()' da resposta converte a resposta JSON em um dicionário Python.
# Ela é usada aqui para imprimir os detalhes da resposta em formato legível.
print(response.json())

```

```python
##Método GET

# Importando a biblioteca 'requests' para fazer requisições HTTP.
import requests

# URL da API que será usada para consultar informações de um usuário específico (usuário com ID 1).
url = 'https://jsonplaceholder.typicode.com/users/1'

# Fazendo uma requisição GET à URL da API para obter informações do usuário.
response = requests.get(url)

# Imprimindo o código de status da resposta HTTP.
print(response.status_code)
# A função 'json()' da resposta converte a resposta JSON em um dicionário Python.
# Ela é usada aqui para imprimir os detalhes da resposta em formato legível.
print(response.json())

```

```python
##Método PUT

# Importando a biblioteca 'requests' para fazer requisições HTTP.
import requests
# Importando o módulo 'json' para lidar com dados em formato JSON.

# URL da API que será usada para atualizar informações de um usuário específico (usuário com ID 1).
url = 'https://jsonplaceholder.typicode.com/users/1'

# Dados atualizados do usuário que serão enviados para a API.
updated_user = {
    'name': 'Novo Nome',
    'username': 'novonome'
}

# Fazendo uma requisição PUT à URL da API, enviando os dados atualizados do usuário em formato JSON.
response = requests.put(url, json=updated_user)

# Imprimindo o código de status da resposta HTTP.
print(response.status_code)
# A função 'json()' da resposta converte a resposta JSON em um dicionário Python.
# Ela é usada aqui para imprimir os detalhes da resposta em formato legível.
print(response.json())

```

```python
##Método delete

# Importando a biblioteca 'requests' para fazer requisições HTTP.
import requests

# URL da API que será usada para excluir informações de um usuário específico (usuário com ID 1).
url = 'https://jsonplaceholder.typicode.com/users/1'

# Fazendo uma requisição DELETE à URL da API para excluir informações do usuário.
response = requests.delete(url)

# Imprimindo o código de status da resposta HTTP.
print(response.status_code)

```

## 1.4.2 Respostas HTTP.

Uma resposta a uma solicitação ou requisição REST precisa conter os seguintes componentes:
* **Linha de status:** Informa ao cliente o resultado (sucesso ou falha) do processamento da
requisição solicitada. As linhas de status são definidas por um código numérico de três dígitos
e uma descrição textual curta, sendo que códigos 2XX indicam sucesso, enquanto que códigos
4XX ou 5XX indicam falha no processamento da requisição;
* **Corpo da resposta:** A grande parte dos métodos retornam dados adicionais além do status,
enviados no corpo da resposta com a representação do recurso processado pelo servidor. Por
exemplo, respostas aos métodos POST e GET retornam a representação do recurso ou indicado
no URL, enquanto uma resposta ao método DELETE retornará um corpo sem conteúdo;
* **Cabeçalho da resposta:** Algumas respostas podem conter dados adicionais enviados no
cabeçalho da resposta a fim de fornecer dados adicionais a respeito do processamento realizado pelo servidor.


## 1.4.3 O Endpoint na prática.

Você acessa os recursos por meio de URLs (Uniform Resource Locators), da mesma forma que, ao acessar um URL em seu navegador, recupera um recurso de informações. Os URLs são acompanhados de um método que especifica como você deseja interagir com o recurso.

Os métodos comuns já mencionados incluem GET (leitura), POST (criação), PUT (atualização) e DELETE (remoção). O endpoint geralmente inclui parâmetros de consulta que especificam mais detalhes sobre a representação do recurso que você deseja ver. Por exemplo, você pode especificar (em um parâmetro de consulta) que deseja limitar a exibição a 5 instâncias do recurso. Como mostrado abaixo.

```python
http://apiserver.com/homes?limit=5&format=json
```

O endpoint mostra o caminho completo para o recurso. No entanto, na documentação, você geralmente separa esse URL em partes mais específicas:

* O caminho base (ou URL base ou host) refere-se ao caminho comum para a API. No exemplo acima, o caminho base é http://apiserver.com.
* O endpoint refere-se ao caminho final do endpoint. No exemplo acima, `/homes`.
* A parte `?limit=5&format=json` do endpoint contém parâmetros de string de consulta para o endpoint.

No exemplo acima, esse ponto de extremidade obteria o recurso "homes" e limitaria o resultado a 5 homes. Ele retornaria a resposta no formato JSON.

Você pode ter vários endpoints que se referem ao mesmo recurso. Aqui está uma variação:

```python
http://apiserver.com/homes/{home id}
```

# 1.5 Autenticação em uma API.

A autenticação de API é um componente essencial quando se trata de segurança e controle de acesso em serviços web. Em um mundo onde as APIs (Interfaces de Programação de Aplicativos) desempenham um papel crucial na comunicação e troca de dados entre diferentes sistemas, a autenticação se torna fundamental para garantir que apenas as entidades autorizadas possam acessar e interagir com os recursos protegidos por uma API.

A autenticação de API envolve verificar a identidade do usuário ou aplicação que está fazendo uma solicitação à API. Isso é especialmente importante para proteger dados sensíveis e prevenir acessos não autorizados. Existem várias estratégias de autenticação de API, cada uma com suas vantagens e cenários de uso específicos:

* **Chave de API (API Key)**: Uma chave de API é um código exclusivo fornecido aos desenvolvedores ou aplicativos que desejam acessar a API. A chave é incluída na solicitação como um parâmetro ou em um cabeçalho HTTP. Embora seja fácil de implementar, a segurança depende do segredo da chave, que pode ser comprometido se não for tratado corretamente.

* **Token de Acesso:** Tokens de acesso são strings geradas que são usadas para autenticar usuários ou aplicativos. Eles são geralmente temporários e podem ser revogados. Os tokens podem ser enviados como parte do cabeçalho de autorização em cada solicitação ou como um parâmetro. Tokens também podem conter informações adicionais, como escopo e data de expiração.

* **OAuth:** OAuth (Open Authorization) é um protocolo de autorização amplamente usado para delegar permissões entre diferentes serviços. Ele permite que um aplicativo acesse recursos em nome de um usuário, sem expor as credenciais do usuário ao aplicativo. É comumente usado para autorizar aplicativos a acessar contas de mídias sociais ou outras plataformas.

## <h1> Exemplo 1: Adquirindo dados por API </h1>

Vamos usar a API pública "Rest Countries" para extrair informações sobre um país específico. Nesta API, podemos obter dados detalhados sobre vários países do mundo.


```python
import requests

# Nome do país que desejamos obter informações.
country_name = 'Canada'

```

Pela documentação da API em https://restcountries.com/#endpoints. A requisição deve seguir:

image.png

```python
# URL da API para acessar informações do país específico.
url = f'https://restcountries.com/v3.1/name/{country_name}'
```

```python
# Fazendo uma requisição GET para a URL da API.
response = requests.get(url)

```

Response = 200, significa que houve retorno de dados.

```python
response
```

A documentação da API também informa tudo o que vem na resposta.

image.png

response.jason() mostra os dados em JSON que vieram na chamada da API

```python
response.json()
```

Caso a requisição seja bem-sucedida, extrairemos as informações do JSON.

```python
# Verificando se a requisição foi bem-sucedida (status 200).
if response.status_code == 200:
    # Convertendo a resposta JSON para uma lista de dicionários Python.
    country_data = response.json()

    # Extraindo informações específicas do país.
    country_info = country_data[0]

    country_name = country_info['name']['common']
    capital = country_info['capital'][0]
    population = country_info['population']
    area = country_info['area']
    region = country_info['region']

    # Exibindo as informações do país.
    print(f"Country: {country_name}")
    print(f"Capital: {capital}")
    print(f"Population: {population}")
    print(f"Area: {area} km²")
    print(f"Region: {region}")
else:
    print("Error accessing the API.")
```

Vamos usar a API pública da NASA para extrair informações sobre um país específico. Nesta API, podemos obter dados, imagens e vídeos detalhados sobre planetas, estrelas e constelações.


1) Entre na URL (https://api.nasa.gov/) e obtenha sua chave de API. Guarde ela com você.

Isso é muito importante para entender como fazer requisição e obter os dados. Toda API possui uma documentação. Na imagem abaixo, a documentação informa a URL e o método GET para obter as informações.

3) Ela também informa os parâmetros que devem conter neste método (data inicial, data final, quantas imagens serão adquiridas, a chave da API e etc) , como na figura abaixo.

image.png

4) A API fornece algum exemplo de query (requisição), como: https://api.nasa.gov/planetary/apod?api_key=DEMO_KEY

Vamos começar importando a lib `requests`, criando a variável da url e chave da api

```python
import requests

# URL base da API da NASA
base_url = "https://api.nasa.gov/planetary/apod"

# Chave de API da NASA
api_key = "SUA_CHAVE"



```

Você pode coolocar os parâmteros da requisição por um dict como o exemplo abaixo

```python
# Parâmetros da solicitação
params = {
    "api_key": api_key,
    "count": 5  # Número de imagens para obter
}

# Fazendo a solicitação GET à API da NASA
response = requests.get(base_url, params=params)
```

Ou você pode colocar como várias strings, como abaixo

```python
# Parâmetros da solicitação na URL, fazendo contrução da string por variáveis
url = f"{base_url}?api_key={api_key}&count=5"
#https://api.nasa.gov/planetary/apod?api_key=api_key&count=5

# Fazendo a solicitação GET à API da NASA
response = requests.get(url)
```

Fazemos um tratamento de erro, caso a resposta não seja 200 (que houve conexão), printa o código de status. Se foi OK, obtemos as informações.

```python

# Verificando se a solicitação foi bem-sucedida
if response.status_code == 200:
    data = response.json()
    for image in data:
        print(f"Data: {image['date']}")
        print(f"Title: {image['title']}")
        print(f"Explanation: {image['explanation']}")
        print(f"URL: {image['url']}")
        print("-" * 50)
else:
    print(f"Erro na solicitação. Código de status: {response.status_code}")
```
