# 1.3 Estrutura de uma página HTML

Para utilizar o Beautiful Soup efetivamente, é importante ter um entendimento básico de como uma página HTML é estruturada. O Beautiful Soup é uma biblioteca Python que facilita a análise e a manipulação de documentos HTML (ou XML). Ele permite que você navegue na árvore de elementos HTML, extraia dados específicos, manipule atributos e conteúdo, e muito mais.

Aqui estão algumas noções básicas sobre páginas HTML que podem ajudar ao trabalhar com o Beautiful Soup:

**1. Estrutura da árvore de elementos:** Uma página HTML é composta por elementos que são organizados em uma estrutura hierárquica de árvore. Cada elemento é representado por uma tag HTML, como `<div>, <p>, <h1>, <a>`, etc. Alguns elementos podem conter outros elementos, criando uma estrutura aninhada.

**2. Atributos:** Cada elemento HTML pode ter atributos que fornecem informações adicionais sobre o elemento. Por exemplo, a tag `<a>` pode ter um atributo href que define o link de destino.

**3. Tags de abertura e fechamento:** A maioria das tags HTML tem uma tag de abertura e uma tag de fechamento correspondente. A tag de abertura começa com < e termina com >, enquanto a tag de fechamento começa com </ e também termina com >. O conteúdo entre as tags é o que é exibido no navegador.

**4. Tags vazias:** Algumas tags, como `<img>` ou `<br>`, são chamadas de tags vazias e não possuem uma tag de fechamento correspondente. Elas geralmente têm atributos para definir seu comportamento.

**5. Classes e IDs:** Atributos de classes e IDs são frequentemente usados para identificar elementos específicos em uma página. Eles são usados para aplicar estilos CSS ou para segmentar elementos usando JavaScript ou frameworks como o Beautiful Soup.

**6. Comentários:** Os comentários em HTML são usados para adicionar notas explicativas que não serão exibidas no navegador. Eles são definidos entre `<!-- e -->.`

**7. Formulários:** Elementos de formulário, como `<input>, <textarea> e <button>`, são usados para interação do usuário e envio de dados.



Segue um exemplo simples de uma página HTML que inclui diversos elementos e atributos:

```python
<!DOCTYPE html>
<html>
<head>
    <title>Página de Exemplo</title>
</head>
<body>
    <header>
        <h1 id="titulo">Bem-vindo ao Meu Site</h1>
        <nav>
            <ul class="menu">
                <li><a href="#">Início</a></li>
                <li><a href="#">Sobre</a></li>
                <li><a href="#">Serviços</a></li>
                <li><a href="#">Contato</a></li>
            </ul>
        </nav>
    </header>

    <main>
        <section id="sobre">
            <h2>Sobre Nós</h2>
            <p>Nossa empresa se dedica a fornecer soluções criativas...</p>
        </section>

        <section id="servicos">
            <h2>Nossos Serviços</h2>
            <ul>
                <li>Web Design</li>
                <li>Desenvolvimento Web</li>
                <li>Marketing Digital</li>
            </ul>
        </section>
    </main>

    <footer>
        <p>&copy; 2023 Meu Site. Todos os direitos reservados.</p>
    </footer>
</body>
</html>
```

**1.** A página começa com a declaração <!DOCTYPE html> indicando que é um documento HTML5.

**2.** Existem várias tags de abertura e fechamento, como `<html>, <head>, <title>, <body>, <header>, <nav>, <ul>, <li>, <section>, <h1>, <h2>, <p>, <footer>`.

**3.** Atributos como id, class e href são usados em vários elementos para identificação, aplicação de estilos CSS e links.

**4.** Há um exemplo de comentário HTML `<!-- Comentário aqui -->`.

**5.** A estrutura inclui um cabeçalho com um título, uma navegação, conteúdo principal com seções sobre e serviços, e um rodapé.

# 1.4 Como obter dados de um Website?

Quando você executa o código para raspagem da Web, uma solicitação é enviada para o URL que você mencionou. Como resposta à solicitação, o servidor envia os dados e permite que você leia a página HTML ou XML. O código, então, analisa a página HTML ou XML, encontra os dados e os extrai.

Para extrair dados usando web scraping com python, você precisa seguir estas etapas básicas:

1. Localizar a URL que você deseja extrair
2. Inspecionar a página
3. Encontrar os dados que você deseja extrair
4. Escrever o código
5. Executar o código e extrair os dados
6. Armazenar os dados no formato necessário




```python
import requests
from bs4 import BeautifulSoup
import re
```

```python
url = 'https://sc.senai.br/'
response = requests.get(url)
```

```python
html_content = response.content
```

```python
# Criação de um objeto Beautiful Soup chamado soup ao analisar algum conteúdo HTML.
soup = BeautifulSoup(html_content, 'html.parser')
```

```python
# Essa linha de código usa o método find_all do objeto Beautiful Soup para procurar todas as ocorrências dos elementos HTML especificados na lista.
text_elements = soup.find_all(['p', 'h1', 'h2', 'h3', 'h4', 'h5', 'h6', 'span'])
```

* 'p': Elementos de parágrafo
* 'h1', 'h2', 'h3', 'h4', 'h5', 'h6': Elementos de cabeçalho de diferentes níveis (de 1 a 6)
* 'span': Elementos de span

```python
#Extraindo os elementos do texto e concatenando em strings
scraped_text = ' '.join(element.get_text() for element in text_elements)
print(scraped_text)
```

1. ' '.join(element.get_text() for element in text_elements): Esta linha de código está utilizando uma expressão geradora (generator expression) para iterar sobre a lista de elementos HTML (text_elements) que você encontrou anteriormente. Para cada elemento na lista, a função get_text() é chamada. Essa função extrai o texto contido dentro do elemento, removendo qualquer marcação HTML. Em seguida, a função join() é usada para concatenar todos esses textos extraídos em uma única string, separados por um espaço em branco.

2. scraped_text: Após a concatenação de todos os textos extraídos, a string resultante é atribuída à variável scraped_text. Esta variável agora contém todo o texto extraído dos elementos HTML.

3. print(scraped_text): Por fim, o conteúdo da variável scraped_text é impresso na saída padrão, geralmente na janela de saída do console. Isso mostrará o texto extraído de todos os elementos HTML especificados na lista text_elements.

```python
# Achando o primeiro parágrafo
first_paragraph = soup.find('p').get_text()
print(first_paragraph)
```

Para verificar a classe relativa ao dado que você quer, no caso o título de um livro, pelo Google Chrome, basta inspecionar o item clicando com o botão direto do mouse em cima dele, como na figura abaixo.

```python
from bs4 import BeautifulSoup
import requests

# URL da página para extrair dados
url = "http://books.toscrape.com/"

# Fazendo uma requisição HTTP para obter o conteúdo da página
response = requests.get(url)
html_content = response.content

# Criar um objeto Beautiful Soup a partir do conteúdo HTML
soup = BeautifulSoup(html_content, 'html.parser')

# Encontrar todos os elementos de produtos
product_elements = soup.find_all('h3')  # H3 contém os títulos dos produtos

for product in product_elements:
    title = product.a.attrs['title']  # Obtém o título do produto

    # Encontrar o elemento de preço
    price_element = product.find_next('p', class_='price_color')
    price = price_element.get_text()  # Obtém o preço do produto

    print("Título:", title)
    print("Preço:", price)
    print("---")
```

1. Importamos as bibliotecas BeautifulSoup e requests.
2. Definimos a URL do site de exemplo "http://books.toscrape.com/".
3. Fazemos uma requisição HTTP usando `requests.get()` para obter o conteúdo HTML da página.
4. Criamos um objeto Beautiful Soup usando o analisador 'html.parser'.
5. Usamos o método `find_all()` para localizar todos os elementos de produto, que são representados pelas tags `<h3>` na página.
6. Iteramos sobre esses elementos e extraímos os títulos dos produtos usando `product.a.attrs['title']`.
7. Imprimimos os títulos dos produtos.

**Jeito 2**

```python
from bs4 import BeautifulSoup
import requests

# URL da página para extrair dados
url = "http://books.toscrape.com/"

# Fazendo uma requisição HTTP para obter o conteúdo da página
response = requests.get(url)
html_content = response.content

# Criar um objeto Beautiful Soup a partir do conteúdo HTML
soup = BeautifulSoup(html_content, 'html.parser')


bookshelf = soup.findAll(
    "li", {"class": "col-xs-6 col-sm-4 col-md-3 col-lg-3"})

for books in bookshelf:


    book_title = books.h3.a["title"]


    book_price = books.findAll("p", {"class": "price_color"})
    price = book_price[0].text.strip()

    print("Título :" + book_title)
    print("Preço :" + price)

```

## 1.5.2 Exemplo 2 - Dados de hostels.

Segue um exemplo, utilziando dados de buscadores de hostels. Para entender como funciona a busca por classes e variáveis pelo Beautiful Soup, veja as imagens abaixo.

image.png

```python
from requests import get
from bs4 import BeautifulSoup

import pandas as pd
import numpy as np

import time
import re

url = 'https://www.hostelworld.com/hostels/Berlin'
response = get(url)

# create soup
soup = BeautifulSoup(response.text, 'html.parser')

# creating individual containers, on each one there's information about one hostel.
hostel_elements = soup.find_all('a', class_='property-card-container property-listing-card')

# how many hostels on the page
len(hostel_elements)
```

image.png

```python
# Iterar sobre os elementos encontrados
for hostel in hostel_elements:
    # Extrair o nome do hostel
    title = hostel.find_next('div', class_='property-name')
    hostel_name = title.span.get_text(strip=True)

    price = hostel.find_next('div', class_='accommodation-price single-row')
    price_hostel = price.strong.get_text(strip=True)

    print("Nome do Hostel:", hostel_name)
    print("Preço:", price_hostel)
```

# 1.6 Questões éticas de Webscrapping.

Dependendo de como é feito, ou seja, quais dados são raspados e com qual propósito, sim o web scraping será considerado ilegal.

Suponhamos que os dados obtidos sejam dados pessoais, o que segundo a LGPD deve acontecer sob circunstâncias bastante claras e específicas. Os usuários aos quais os dados se referem, saberão para quais fins eles serão utilizados? Poderão solicitar a anonimização deles? Até mesmo sabem que seus dados estão de posse de alguém ou uma empresa diferente daquela para a qual foram originalmente fornecidos?

Neste caso, ainda que os dados sejam públicos e tenham sido colhidos em uma rede social, o usuário ao qual eles estão associados, consentiu em fornecê-los exclusivamente no escopo da rede e para os devidos usos, sob os termos da sua prestação de serviço e sua política de privacidade.

Não foi dado consentimento explícito de utilização fora dela, nem tampouco por parte de terceiros.

Mesmo quando não se trata de dados pessoais e sob a regulação da LGPD, há implicações de ordem moral, ética, profissional e mercadológicas, que precisam ser observadas:

* **Patrimônio digital** – colher dados em sites diversos, dependendo de quais dados são colhidos, com qual propósito, significa apropriar-se do patrimônio digital constituído pelos esforços e recursos de terceiros. Especialmente quando se obtém lucro com tal prática, o qual não ocorreria sem o web scraping e dependendo das circunstâncias, pode-se ter uma situação de infração a propriedade intelectual;

* **Legislação local** – no Brasil há ainda pouquíssima legislação relativa ao ambiente digital, no entanto, dependendo do país de origem do site, pode-se estar infringindo leis locais pelas quais o site é protegido;

* **Termos de serviço** – além da política de privacidade a qual todo site que lide com dados de terceiros precisa ter, há vezes em que há termos de serviços que determinam o escopo de utilização da informação contida, bem como a prestação de serviços associada. Ao burlar ou ignorar tais termos, pode-se estar cometendo um delito previsto em lei;

* **Arquivo robots.txt** – para além da política de privacidade e dos termos de serviço, por convenção, por compromisso ético e profissional, por bom tom e respeito, o bot de web scraping deve atuar com base no arquivo robots.txt dos sites que varrer;

* **Nome do robô** – decorrência direta do item acima, é conveniente e consiste de boa prática, que o scrapper tenha uma identificação conhecida, tal como os bots mais conhecidos, como o Googlebot ou o Bingbot;

* **Desempenho** – a depender do site varrido e de suas características técnicas, do volume de dados e de quais são “raspados”, o web scraping pode funcionar como um ataque ou tentativa de invasão, podendo inclusive prejudicar o site alvo, afetando o seu desempenho;

* **Copyright** – tal como no caso de dados pessoais e dados sensíveis que podem ser classificados como propriedade intelectual, ainda há aqueles que são protegidos por leis de direito autoral e que da mesma forma não podem ser coletados, nem usados sem as devidas permissões.
