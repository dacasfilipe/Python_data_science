```python
from google.colab import drive #Importa o pacote google.colab e a função drive do Google Colab.
drive.mount('/content/drive', force_remount=True) #monta o drive
```

```python
linhas=[]
arquivo = open('/content/drive/MyDrive/Colab Notebooks/mailbox.txt', 'r') #abre arquivo na localização indicada
for linha in arquivo:
  if linha.startswith('From '): linhas.append(linha)
arquivo.close() #fecha arquivo
linhas
```

```python
listagem_emails = []
for linha in linhas:
  partes = linha.split(' ')
  email = partes[1].strip()
  listagem_emails.append(email)
  #print(email)
listagem_emails.sort()
lista_exclusiva = set(listagem_emails)
print(lista_exclusiva)
```

```python
saida = open('/content/drive/MyDrive/Colab Notebooks/output.txt', 'w')  #Veja que o argumento 'w' abre o arquivo no modo gravação (write)
for linha in lista_exclusiva: #estrutura de repetição para gerar 20 linhas no arquiv
  saida.write(f'{linha}\n') #utiliza o método write para gravar cada linha do arquivo
saida.close() #fecha o arquivo
```
