# 1 - Normalização

Em um projeto de banco de dados, os principais objetivos são preservar a informação e manter uma redundância mínima.

Assim, explica-se o uso de relacionamentos, definição e adoção de tipos de dados adequados para cada domínio de valores dos atributos, chaves primárias e estrangeiras, atendimento à cardinalidade mínima e máxima, integridade referencial, etc.

Normalizar um banco de dados exige implementar uma metodologia formal para teste e melhoria dos esquemas de relação.

Além disso, é fundamental que o modelo de dados seja projetado observando-se aspectos de qualidade no armazenamento, recuperação e processamento dos dados. Problemas e anomalias de atualização de dados (inclusão exclusão e alteração), dependências parciais com relação às chaves entre duas ou mais tabelas, redundância de dados, entre outros,  devem ser eliminados a fim de não comprometer a integridade dos dados.

A normalização de dados pode ser considerada um processo para analisar os esquemas de relação de dados com base em suas dependências funcionais e chaves primárias para conseguir as minimizações dos problemas listados acima.
A normalização é uma das técnicas que permite permite depurar um projeto de banco de dados, através da identificação de inconsistências, informações em duplicidade, dependências funcionais mal resolvidas, etc.

“O processo de normalização leva um esquema de relação por uma série de testes para certificar se ele satisfaz certa forma normal” (Elsmari, Navathe, p 348, 2011.)

Portanto...
Normalização é o processo de organização de dados em um banco de dados.
Isso inclui a criação de tabelas e o estabelecimento de relações entre essas tabelas de acordo com as regras projetadas para proteger os dados e tornar o banco de dados mais flexível, eliminando a redundância e a dependência inconsistente.

**Formas normais**

Após a definição do modelo de dados, utiliza-se o processo de normalização para a construção de uma síntese dos dados, cujo objetivo é definir entidades e relacionamentos estáveis e adequadas para sua implementação em um modelo físico do banco de dados.

As regras para normalização do banco de dados são denominadas formas normais. À medida que os níveis da FN aumentam, também aumentam a complexidade e o custo em atender às formas e à conformidade de cada nível.

Realizar a normalização de um banco de dados é, após definir as dependências funcionais entre os atributos de dados, desenvolver uma metodologia formal para teste e melhorias dos esquemas de relação.

**FN1, originalmente 1NF**

Uma relação está na primeira forma normal se todos os seus atributos são monovalorados e atômicos. Ou seja, todos os atributos devem ser simples e indivisíveis. Em determinadas situações, podemos encontrar dados repetidos, com a ocorrência de duplicidades de valores relacionados à mesma chave primária.

• Eliminar grupos repetidos em tabelas individuais.

• Crie uma tabela separada para cada conjunto de dados relacionados.

• Identifique cada conjunto de dados relacionados com uma chave primária.

**FN3, originalmente 3NF**

Uma relação está na terceira forma normal se:

1. A relação estiver na segunda forma normal (FN2) e

2. Todos os atributos dependem não transitivamente de toda a chave primária.

**Resposta das atividades**

As tabelas do banco de dados Geografia podem ser criadas a partir dos scripts disponiveis no tópico da semana 1.

Lembrando que nas tabelas temos problemas de normalização e referência à chave primária da tabela regiao, pela chave estrangeira da tabela estado.

Para ajustar o banco de dados Geografia, deverão ser realizados os seguintes passos:

1.   Eliminar o problema de F1N do banco rios_brasileiros. O atributo *estados* da tabela *rios_brasileiros* não atende a F1N, pois contém uma lista de diversos estados. E como na cardinalidade, um rio pode passar por diversos estados (n) e um estado pode conter diversos rios (n), criaremos a tabela associativa estado_rios, para eliminar esse problema.

Para eliminar esse problema, criamos a tabela associativa 1-n estado_rios.

Script SQL para criar a tabela associativa *estado_rio*, inserindo os atributos e as respectivas chaves estrangeiras relacionando às tabelas rios_brasileiros e estados.
Isso eliminará duas situações problemas:

```
CREATE TABLE `geografia`.`estado_rios` (
  `id` INT UNSIGNED NOT NULL AUTO_INCREMENT,
  `id_rio` INT NOT NULL,
  `id_estado` INT NULL,
  PRIMARY KEY (`id`),
  INDEX `id_estado_idx` (`id_estado` ASC) VISIBLE,
  INDEX `id_rio_idx` (`id_rio` ASC) VISIBLE,
  CONSTRAINT `id_estado`
    FOREIGN KEY (`id_estado`)
    REFERENCES `geografia`.`estado` (`id`)
    ON DELETE NO ACTION
    ON UPDATE NO ACTION,
  CONSTRAINT `id_rio`
    FOREIGN KEY (`id_rio`)
    REFERENCES `geografia`.`rios_brasileiros` (`id`)
    ON DELETE NO ACTION
    ON UPDATE NO ACTION);
```

*Observar que foi criado um atributo id como PK da tabela, mas também pode-se criar uma PK composta, com os campos id_rio e id_estado e que também serão FK, referenciando os respectivos id's das tabelas rios_brasileiros e estados.*

Ele deve ser uma FK fazendo referência ao ida da tabela regiao. Mas está como varchar, com o nome da regiao ao invés do id. A solução é alterar o tipo de dado desse atributo para INT e criar o relacionamento ao ida da tabela regiao.
Um problema que pode ser gerado é que o atributo contém texto, mas se for alterado para tipo int, o SGBD pode acusar erro de conversão. Uma solução é criar um novo atributo, inserir os valores referentes a cada regiao e somente depois, excluir o atributo regiao original.

Renomear o atributo nome para regiao e criar o relacionamento com a tabela regiao.
