```python
df = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/wine.data', sep=',',index_col=False, names=['Type', 'Alcohol', 'Malic acid', 'Ash',
                                                   'Alcalinity of ash', 'Magnesium',
                                                   'Total phenols', 'Flavanoids', 'Nonflavanoid phenols',
                                                   'Proanthocyanins', 'Color intensity',
                                                   'Hue', 'OD280/OD315 of diluted wines',
                                                   'Proline']) #lê arquivo .csv separado por ';'
df
```

## Quantidade de vinhos por tipo

```python
df['Type'].value_counts()
```

## Maior e menor teor alcoólico dos vinhos

```python
biggest_alcohol = df['Alcohol'].max()
lowest_alcohol  =  df['Alcohol'].min()
print(f'Biggest malic acid: {biggest_alcohol}')
print(f'Lower malic acid: {lowest_alcohol}')
```

## Média de Teor alcoólico por Tipo de Vinho

```python
mean_alcohol_wine_type_one = df[df['Type'] == 1]['Alcohol'].mean()
mean_alcohol_wine_type_two = df[df['Type'] == 2]['Alcohol'].mean()
mean_alcohol_wine_type_three = df[df['Type'] == 3]['Alcohol'].mean()
print(f'Mean alcohol - Wine type one: {mean_alcohol_wine_type_one}')
print(f'Mean alcohol - Wine type two: {mean_alcohol_wine_type_two}')
print(f'Mean alcohol - Wine type three: {mean_alcohol_wine_type_three}')
```
