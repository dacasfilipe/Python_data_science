**2.1 - Conceitos**

Banco de dados é um espaço físico onde os dados são armazenados, recuperados e processados por sistemas computacionais.

Esses dados são organizados e **relacionam-se** entre si, sendo definidos e manipulados através de instruções em uma linguagem própria de banco de dados.

O relacionamento existe para permitir a construção de informações a partir do cruzamento dos dados entre duas ou mais tabelas. Tabelas que não possuem relação com nenhuma outra, não devem fazer parte de um banco de dados.

Sobre banco de dados **relacionais**, são conjuntos de dados organizados em estruturas denominadas de TABELAS, e as operações que as utilizam são realizadas por linguagens que manipulam esses dados.

Os principais elementos de um modelo relacional são: entidades ou tabelas, atributos e seus tipos de dados, relacionamentos, chaves e cardinalidade.

**2.2 - Entidade e relacionamento**

No  modelo relacional, o objeto básico, essencial à estrutura lógica, é a **entidade**.

Entidades são aspectos que representam algo presente no mundo real cuja existência é independente de outras entidades. Ainda, uma entidade corresponde a qualquer coisa desse mundo real sobre a qual desejamos armazenar algum tipo de dado.

Conforme  Elmasri e Navathe, “uma entidade pode ser um objeto com existência física (um carro, aluno, livro, município, etc) ou um objeto com existência conceitual (empresa, cargo, curso, etc)”  (Elmasri, Navathe, p. 135, 2011)

Como é um objeto que pode ser representado e tem significado na vida real, as entidades possuem características específicas que lhes dão sentido e materialidade, mesmo que tendo um aspecto conceitual. Pois uma aluno é uma pessoa, portanto, possui nome, endereço, altura, sexo, data de nascimento, altura, peso, entre outros.

Teorey e colegas descrevem uma entidade como “principais objetos de dados sobre os quais informações devem ser coletadas; elas normalmente representam uma pessoa, lugar, coisa ou evento de interesse informativo” (Teorey et al, pag 15, 2014).

Outro exemplo, um carro possui cor, placa, quilometragem, ano de fabricação, modelo, fabricante, e assim por diante.

Outros exemplos do mundo real:
Entidades que representam pessoas, como alunos, professores, administrativo, etc. E que representam objetos materiais ou abstratos, como UC, notas,salas, currículo, histórico escolar, etc.



**2.3 - Cardinalidades**

Um relacionamento possui a indicação de uma cardinalidade entre as ocorrências das entidades relacionadas.
Então, cardinalidade indica o número máximo e mínimo de relacionamentos que podem ocorrer entre tabelas.
*   Cardinalidade mínima
*   Cardinalidade máxima

E os tipos de cardinalidade
*   Um-para-um;
*   Um-para-muitos;
*   Muitos-para-muitos

**Cardinalidade mínima**

Relacionamento obrigatório ou não.

Exemplo: relacionamento entre país e UF
Um país pode ser cadastrado sem nenhuma UF vinculada, mas uma UF está vinculada a UM único país.

Então na cardinalidade mínima...
*   Uma país pode não possuir UF
*   Uma UF pertence a UM único país.
image.png

**Cardinalidade máxima**

Define a quantidade máxima de ocorrência entre as entidades de um relacionamento.
*   Um país pode ter várias UF vinculadas.
*   E a UF pode ter no máximo UM país vinculado.

image.png

**Cardinalidade - Um-para-Um**

Quando uma ocorrência da entidade A se associa com apenas uma ocorrência da entidade B.
Assim como uma ocorrência da entidade B se associa com apenas uma ocorrência da entidade A.
Neste caso, temos um relacionamento com cardinalidade um-para-um (1:1).

**Cardinalidade - Um-para-Muitos**

É a cardinalidade mais comum.
Quando uma ocorrência da entidade A se associa a mais de uma ocorrência da entidade B.
E uma ocorrência da entidade B se associa a uma e somente uma ocorrência da entidade A.

**Cardinalidade Muitos-para-Muitos**

Quando uma ocorrência da entidade A pode se associar a mais de uma ocorrência da entidade B.
E uma ocorrência da entidade B pode se associar a mais de uma ocorrência da entidade A

**2.4 - Atributos**

As características de uma entidade são denominadas de atributos, representadas por colunas nas tabelas  e possuem um determinado valor, que são os dados armazenados, que representam então suas características presente no mundo real.

Esses atributos ocorrem em diversos formatos no modelo relacional: simples ou composto, de valor único ou multivalorado, armazenado ou derivados. Vamos detalhar cada um deles a seguir:

**Atributos simples ou compostos**

Um atributo **simples** é aquele que não pode ser dividido. É o denominado atributo atômico, por conter um valor básico que, se dividido, perde o sentido e a relevância. Por exemplo, Sexo, CPF de uma entidade Pessoa. Placa, combustível, ano de fabricação, de uma entidade Frota.

Um atributo **composto** é aquele que permite sua divisão, sendo que suas partes contêm significado se tratados individualmente.
Por exemplo, nome de um empregado, caso seja relevante termos uma relação por sobrenome, além do nome. Endereço de um empregado, caso seja relevante, e necessário, tratarmos separadamente logradouro, bairro, município, UF, CEP e outros. E geralmente, mas não é regra, que o endereço seja dividido em atributos menores, mantendo a relevância de cada um deles.

**Atributos de valor único ou multivalorado**

Um atributo de valor único, ou monovalorado, é um atributo que possui um único valor para aquela entidade. Por exemplo, o atributo nome é único para uma pessoa, assim como CPF, endereço, data de nascimento. Porém, essa mesma entidade pode ter ocorrência onde um atributo possui mais de um valor para aquela entidade. Uma pessoa pode ter dois ou mais telefones, um aluno pode ter registrado diversas notas para uma mesma unidade curricular.

**Atributos armazenados ou derivados**

Um atributo **derivado** é quando, mesmo sendo relevante, ele pode ser obtido através de processamento em outros atributos ou até mesmo através de relacionamentos com outras entidades. Um exemplo bem clássico é a idade de uma pessoa. Armazenar esse tipo de atributo pode gerar problemas de atualização, pois a cada aniversário esse valor irá mudar. E, podemos obter a idade de uma pessoa através de rotinas computacionais que calculam a idade a partir da data de nascimento.

Outro exemplo de atributo derivado é um quantitativo gerado através de contadores ou acumuladores. No caso do contador, um exemplo é obtermos a quantidade de empregados de determinado setor da empresa. Para isso, podemos gerar um processamento que efetua a contagem de quantas pessoas estão cadastradas naquele setor. No caso de acumulador, por exemplo, o total de salário pago aos empregados do mesmo setor da empresa. Para isso, executamos uma ação que irá somar os salários de cada empregado para obtermos esse resultado.

E, como o próprio nome diz, um atributo **armazenado** é aquele que está salvo no banco de dados e para obter o mesmo, precisamos realizar uma consulta ao banco.

**2.5 - Tipos de Dados e Domínio**

Representa o conjunto de valores admissíveis de um atributo (coluna) de uma tabela.

Por exemplo:
*   MATRÍCULA: 3 dígitos.
*   NOME: máximo de 50 caracteres alfa.
*   DATA NASC: data válida.
*   ENDEREÇO: máximo 100 caracteres alfanumérico e símbolos.
*   INGRESSO: inteiro com 5 dígito


**Domínio – Restrições (constraint)**

Restrições são determinadas a partir da definição de determinando domínios de valores para cada coluna de uma tabela e também os valores que uma coluna de uma tabela pode ter.

Restrições permitem:
*   Verificar os valores inseridos em um banco e dados;
*   Testar consultas para garantir que as comparações tenham sentido;
*   Em geral, o domínio é especificado por tipos primitivos de dados, tais como integer, float, char, date, time, etc...
*   Uma restrição muito utilizada na definição de atributos válidos em uma tabela é o NOT NULL, ou seja, não é permitido valor nulo para aquela atributo.

**Chave Primária**

Em toda tabela haverá uma coluna, ou conjunto de colunas, onde o valor deve ser único, não será NULL tampouco repetirá.
Esta coluna, ou conjunto de colunas, é denominada chave primária.

Na tabela abaixo, o atributo MATRÍCULA é chave primária.

image.png

**Chave primária composta**

É possível concatenar mais de uma coluna para compor a chave primária.
É importante observar que, esta concatenação NÃO pode resultar em dois valores iguais em hipótese alguma.

Na tabela Turmas, os atributos UC e ALUNO formam a chave primária, pois a combinação desses dois valores não poderá ser repetido.

image.png

**Chave estrangeira**

Uma chave estrangeira define um relacionamento entre as tabelas A e B.
Em uma tabela B, uma chave estrangeira sempre será chave primária na tabela A.
Utilizando o mesmo exemplo das tabelas aluno, UC e Turmas, os campos MATRÍCULA da tabela Aluno é única para cada aluno, portanto ela foi definida como chave primária desta tabela.

O campo UC também é chave primária na tabela UC. Na tabela Turmas, percebe-se a ocorrência dos valores cujos campos são chave primária em outra tabela. Nessa, portanto, os campos UC e Aluno são chaves estrangeiras, pois estão definidas em outras tabelas, mas são utilizadas na tabela Turmas para identificar cada aluno e cada UC.
